```python
mean_values = df.mean()
new_row_mean = pd.DataFrame([mean_values], columns=df.columns)
df = pd.concat([df, new_row_mean], ignore_index=True)
display(df.tail())
```

```python


df.loc[df.index[-1], 'A'] = np.nan
display(df.tail())
```

```python
df.dropna(inplace =True)
display(df.tail())
```

```python
df['target']=df['target'].astype(float)
df.dtypes
```

```python
df=df.rename(columns = {'A' : 'Area'})
df=df.rename(columns = {'P' : 'Perimeter'})
df=df.rename(columns = {'C' : 'Compactness'})
df=df.rename(columns = {'LK' : 'Length of Kernel'})
df=df.rename(columns = {'WK' : 'Width of Kernel'})
df=df.rename(columns = {'A_Coef' : 'Asymmetry of Kernel'})
df=df.rename(columns = {'LKG' : 'Length of Kernel Groove'})
```

```python
# Select numeric columns  # Multi column box plot
data = df[[
    "Area",
    "Length of Kernel Groove",
    "Asymmetry of Kernel",
]]

plt.figure(figsize=(10, 6))

# Create box plot
box = plt.boxplot(
    data.values,
    tick_labels=data.columns,
    patch_artist=True,
    medianprops=dict(color="red", linewidth=2),
    boxprops=dict(facecolor="lightblue"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black")
)

# Labels and title
plt.xlabel("Area vs Lenght of KG vs Asymmetry of Kernel")
plt.ylabel("Values")
plt.title("BoxPlots")

# Calculate quartiles
Q1 = data.quantile(0.25)
Q2 = data.quantile(0.50)
Q3 = data.quantile(0.75)

# Annotate Q1, Median, Q3 outside the box to reduce congestion
for i, col in enumerate(data.columns, start=1):
    plt.text(i, Q1[col] - 0.2, f"Q1={Q1[col]:.2f}",
             fontsize=9, color="blue", ha="center")
    plt.text(i, Q2[col], f"Median={Q2[col]:.2f}",
             fontsize=9, color="red", ha="center", va="bottom")
    plt.text(i, Q3[col] + 0.2, f"Q3={Q3[col]:.2f}",
             fontsize=9, color="green", ha="center")

# Optional: horizontal grid lines for clarity
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
